# Date Conversion and Feature Engineering

```python
# Convert date columns to datetime
date_cols = ['booking_date', 'check_in_date', 'check_out_date', 'travel_date']
for col in date_cols:
    data[col] = pd.to_datetime(data[col], errors='coerce')
```

```python
# Feature Engineering

# 1. Length of Stay (in days)
# For cancelled/failed bookings, this will be NaN
data['length_of_stay'] = (data['check_out_date'] - data['check_in_date']).dt.days

# 2. Lead Time (days between booking and check-in)
# For cancelled/failed bookings, this will be NaN
data['lead_time'] = (data['check_in_date'] - data['booking_date']).dt.days

# 3. Booking Month (for seasonality analysis)
data['booking_month'] = data['booking_date'].dt.month

# 4. Cancellation Flag
# 1 = Cancelled, 0 = Confirmed/Other
data['is_cancelled'] = data['booking_status'].apply(lambda x: 1 if x == 'Cancelled' else 0)

# 5. Failed Booking Flag 
data['is_failed'] = data['booking_status'].apply(lambda x: 1 if x == 'Failed' else 0)
```

# Exploratory Data Analysis

```python
# Booking Distribution by Channel
# Understanding which channels are most popular for bookings

channel_counts = data['channel_of_booking'].value_counts()

sns.countplot(data=data, x='channel_of_booking', order=channel_counts.index)
plt.title("Number of Bookings by Channel")
plt.ylabel("Number of Bookings")
plt.xlabel("Channel")
plt.xticks(rotation=45)
plt.show()

# Display percentages
print(round(channel_counts / len(data) * 100,2))
```

Most bookings are through the Web channel, followed by Android and iOS. This indicates that customers prefer online booking platforms over offline or corporate channels. Strategic promotions on these top channels can help maximize bookings.

```python
# Revenue by Channel
# Total revenue contribution by channel

revenue_channel = data.groupby('channel_of_booking')['selling_price'].sum().sort_values(ascending=False)

revenue_channel.plot(kind='bar', color='skyblue')
plt.title("Total Revenue by Channel")
plt.ylabel("Revenue")
plt.xlabel("Channel")
plt.xticks(rotation=45)
plt.show()
```

While Web generates the highest number of bookings, revenue distribution shows that certain channels may have higher average booking values. This helps identify high-value channels for targeted marketing.

```python
# Star Rating vs Selling Price

sns.boxplot(data=data, x='star_rating', y='selling_price')
plt.title("Booking Value by Star Rating")
plt.xlabel("Star Rating")
plt.ylabel("Selling Price")
plt.show()
```

Higher star-rated properties generally have higher booking values. Focusing marketing on 4–5 star properties could increase profitability.

```python
#Overall Cancellation Rate
cancel_rate = data['is_cancelled'].mean() * 100
print("Overall Cancellation Rate: {:.2f}%".format(cancel_rate))
```

```python
#Cancellation Rate by Channel
cancel_channel = data.groupby('channel_of_booking')['is_cancelled'].mean() * 100
cancel_channel.plot(kind='bar', color='salmon')
plt.title("Cancellation Rate by Channel")
plt.ylabel("Cancellation Rate (%)")
plt.xlabel("Channel")
plt.xticks(rotation=45)
plt.show()
```

Channel like Android followed by Web and iOS may have higher cancellation rates. Understanding why some channels are more prone to cancellations helps design targeted strategies to reduce them.

# Root Cause Analysis

```python
# Lead Time vs Cancellation
# Check if bookings made far in advance are more likely to be cancelled

sns.boxplot(data=data, x='is_cancelled', y='lead_time')
plt.title("Lead Time vs Cancellation")
plt.xlabel("Cancelled (1=Yes, 0=No)")
plt.ylabel("Lead Time (days)")
plt.show()

# Average lead time for cancelled vs confirmed bookings
avg_lead = data.groupby('is_cancelled')['lead_time'].mean()
print(avg_lead)
```

Average lead time is similar for cancelled (30.21 days) and confirmed bookings (30.39 days), suggesting that in this dataset, lead time alone is not a significant driver of cancellations. Other factors like room type, booking channel, and seasonal trends are likely influencing cancellation behavior.

```python
# Cancellation Rate by Booking Channel
channel_cancel = data.groupby('channel_of_booking')['is_cancelled'].mean() * 100

# Plot
channel_cancel.plot(kind='bar', color='skyblue')
plt.title("Cancellation Rate by Booking Channel (%)")
plt.ylabel("Cancellation Rate (%)")
plt.xlabel("Channel")
plt.xticks(rotation=45)
plt.show()

# Display exact percentages
channel_cancel.round(2)
```

Cancellation rates are similar across all booking channels (Android 20%, Web 20.3%, iOS 20.4%). This indicates that the channel itself does not significantly influence cancellations. Business interventions to reduce cancellations may be more effective if focused on room types, star ratings, or seasonal trends rather than the booking platform.

```python
# Cancellation Rate by Room Type
room_cancel = data.groupby('room_type')['is_cancelled'].mean() * 100

room_cancel.plot(kind='bar', color='teal')
plt.title("Cancellation Rate by Room Type (%)")
plt.ylabel("Cancellation Rate (%)")
plt.xlabel("Room Type")
plt.show()

# Displaying exact percentages
room_cancel.round(2)
```

Standard rooms have the highest cancellation rate at 23.3%, while Deluxe and Suite rooms have lower rates (~16–18%). This indicates that guests booking budget rooms are more likely to cancel. Strategies to reduce cancellations should prioritize Standard room bookings, such as offering partial prepayment or incentives to confirm the stay.

```python
# Cancellation Rate by Star Rating
star_cancel = data.groupby('star_rating')['is_cancelled'].mean() * 100

star_cancel.plot(kind='bar', color='orange')
plt.title("Cancellation Rate by Star Rating (%)")
plt.ylabel("Cancellation Rate (%)")
plt.xlabel("Star Rating")
plt.show()

# Displaying exact percentages
star_cancel.round(2)
```

Cancellation rates by star rating are fairly consistent across 2–4 stars (~20%), but slightly higher for 5-star properties (21.3%). This suggests that while cancellations are fairly uniform, premium bookings carry slightly higher risk. Business strategies to reduce cancellations could include partial prepayment, early check-in incentives, or reminders targeted at high-value 5-star bookings.

```python
# Cancellation rate by stay type
stay_cancel = data.groupby('stay_type')['is_cancelled'].mean() * 100
stay_cancel.round(2)
```

Cancellation rates are similar for Business (20.05%) and Leisure (20.35%) stays. This indicates that both segments exhibit comparable cancellation behavior, so interventions to reduce cancellations can be applied broadly rather than targeting a specific stay type.

```python
# Cancellation rate by coupon usage
coupon_cancel = data.groupby('Coupon USed?')['is_cancelled'].mean() * 100
coupon_cancel.round(2)
```

Bookings with coupons show a slightly higher cancellation rate (20.8%) compared to those without coupons (20.1%). Although the difference is small, it indicates that discount-seeking customers are marginally more likely to cancel. Businesses could consider partial prepayment or confirmation incentives for bookings made with coupons.

```python
# Monthly Booking Trend
# Count number of bookings per booking month
monthly_bookings = data.groupby('booking_month')['customer_id'].count()

# Plot bookings per month
monthly_bookings.plot(marker='o', linestyle='-', color='blue')
plt.title("Monthly Booking Trend")
plt.xlabel("Booking Month")
plt.ylabel("Number of Bookings")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()

# Displaying exact numbers
monthly_bookings
```

April stands out as a peak month with 4494 bookings, almost double the usual monthly volume (~2300).
This could indicate a holiday, festival, or seasonal demand spike.

```python
# Monthly Cancellation Rate
monthly_cancel = data.groupby('booking_month')['is_cancelled'].mean() * 100

# Plot cancellation rate per month
monthly_cancel.plot(marker='o', linestyle='-', color='red')
plt.title("Monthly Cancellation Rate (%)")
plt.xlabel("Booking Month")
plt.ylabel("Cancellation Rate (%)")
plt.xticks(range(1,13))
plt.grid(True)
plt.show()

# Display exact percentages
monthly_cancel.round(2)
```

Highest cancellation rates in July (30.33%), August (28.77%), and December (25.76%).
These may correspond to peak travel periods or holiday months, where customers make bookings early but sometimes cancel due to last-minute plan changes.

```python
# Combined Bookings and Cancellations
fig, ax1 = plt.subplots(figsize=(10,5))

ax1.set_xlabel("Booking Month")
ax1.set_ylabel("Number of Bookings", color="blue")
ax1.plot(monthly_bookings.index, monthly_bookings.values, color="blue", marker='o', label="Bookings")
ax1.tick_params(axis='y', labelcolor="blue")

ax2 = ax1.twinx()
ax2.set_ylabel("Cancellation Rate (%)", color="red")
ax2.plot(monthly_cancel.index, monthly_cancel.values, color="red", marker='x', label="Cancellation Rate")
ax2.tick_params(axis='y', labelcolor="red")

plt.title("Monthly Bookings vs Cancellation Rate")
fig.tight_layout()
plt.show()
```
